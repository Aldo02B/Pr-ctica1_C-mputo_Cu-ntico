### 3.2 Código Ejemplo 3: mostrar el circuito cuántico
Para mostrar el circuito cuántico se ejecutó un código nuevo con el que pueda acceder a las funcionaldiades del framework Cisq. En los próximos ejemplos se muestra el código original y el resultado. 
>import cirq
>import numpy as np
>import matplotlib.pyplot as plt
>from mpl_toolkits.mplot3d import Axes3D
># Crear un qubit.
>qubit = cirq.GridQubit(0, 0)
># Crear el circuito en Cirq.
>circuit = cirq.Circuit()
># Aplicar la compuerta Hadamard.
>circuit.append(cirq.H(qubit))
># Simular el circuito.
>simulator = cirq.Simulator()
>result = simulator.simulate(circuit)
># Obtener el estado final del qubit.
>state = result.final_state_vector
>bloch_vector = cirq.bloch_vector_from_state_vector(state, 0)
># Función para graficar el vector de Bloch.
>def plot_bloch_vector(bloch_vector):
>    # Crear una figura y un eje 3D para la esfera de Bloch.
>    fig = plt.figure()
>    ax = fig.add_subplot(111, projection='3d')
>    ax.set_xlim([-1,1])
>    ax.set_ylim([-1,1])
>   ax.set_zlim([-1,1])
>   ax.set_xlabel('X')
>   ax.set_ylabel('Y')
>   ax.set_zlabel('Z')
>    # Dibujar el vector
>    ax.quiver(0, 0, 0, bloch_vector[0], bloch_vector[1], bloch_vector[2], color='blue', length=1, normalize=True)
>    # Dibujar la esfera de Bloch.
>    phi, theta = np.mgrid[0:2*np.pi:100j, 0:np.pi:50j]
>    x = np.sin(theta) * np.cos(phi)
>    y = np.sin(theta) * np.sin(phi)
>    z = np.cos(theta)
>   # Dibujar la esfera como una malla de alambre.
>    ax.plot_wireframe(x, y, z, color='black', alpha=0.1)
>    ax.grid(False)
>    # Establecer el fondo como blanco.
>    ax.set_facecolor('white')
>    # Ocultar los ejes.
>    ax.set_xticks([])
>    ax.set_yticks([])
>    ax.set_zticks([])
>    # Asegurarse de que la relación de aspecto sea igual para que la esfera sea circular.
>    ax.set_box_aspect([1,1,1])
>    # Mostrar la figura.
>    plt.show()
># Graficar el vector de Bloch.
>plot_bloch_vector(bloch_vector)





### 3.3 Código Ejemplo 5: mostrar el circuito cuántico
Para mostrar el circuito cuántico se ejecutó un código nuevo con el que pueda acceder a las funcionaldiades del framework Cisq. En los próximos ejemplos se muestra el código original y el resultado. 

>import cirq
>import matplotlib.pyplot as plt
>from cirq import Simulator, H, measure
># Crear dos qubits
>q0, q1 = cirq.LineQubit.range(2)
># Crear el circuito en Cirq
>circuit = cirq.Circuit()
># Aplicar la compuerta Hadamard a ambos qubits
>circuit.append([H(q0), H(q1)])
># Añadir mediciones
>circuit.append([measure(q0, key='m0'), measure(q1, key='m1')])
># Dibujar el circuito
>print(circuit)
># Ejecutar la simulación
>simulator = Simulator()
>result = simulator.run(circuit, repetitions=1000)
># Obtener resultados y graficar el histograma
>cirq.plot_state_histogram(result, plt.subplot())
>plt.show()
